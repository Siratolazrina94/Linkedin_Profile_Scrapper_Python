```python
link = 'https://www.linkedin.com/in/siratolazrina/'
browser.get(link)
```

```python
SCROLL_PAUSE_TIME = 5

#Get scroll height
last_height = browser.execute_script("return document.body.scrollHeight")

for i in range(3):
    #Scroll down to bottom
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    
    # Calculate new scroll height and compare with text scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
src = browser.page_source
soup = BeautifulSoup(src, 'lxml')
soup
```

```python
name_div = soup.find('div',{'class': 'pv-text-details__left-panel mr5'})
name_div
```

```python
name_loc = name_div.find_all('ul')
name = name_div.find('h1').get_text().strip()
name
```

```python
loc = name_div.find('span').get_text().strip()
loc
```

```python
Profile_Title = name_div.find_all('div')[1].get_text().strip()
Profile_Title
```

```python
connection_ = soup.find('ul', {'class' : 'pv-top-card--list pv-top-card--list-bullet display-flex pb1'})
connection_
connection = connection_.find('li').get_text().strip()
connection
```

```python
info = []
info.append(link)
info.append(name)
info.append(Profile_Title)
info.append(loc)
info.append(connection)
info
```
